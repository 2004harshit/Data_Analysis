## Q1 finding average rating

```python
data
```

```python
print("Rating of Apps :")
print(data['Rating'])
print('Average rating :')
print(data['Rating'].mean())
```

```python
# finding rating manually
def calculate_avg_rating(data,sum =0):
    for i in data['Rating']:
        sum+=i
    n = data['Rating'].size
    avg=sum/n
    return avg

print(calculate_avg_rating(data,0))
```

## Q2 How many apps with rating 5 ?

```python
def no_5_rating_apps(data,count=0):
    for i in data['Rating']:
        if i==5:
            count+=1
    return count


no_5_rating_apps(data)
```

```python
paid=0
free=0

for i in df['Type']:
    if i =='Free':
        free+=1
    elif i=='Paid':
        paid+=1
   
print(paid,free)
```

```python
paid_percent= (paid/(paid+free))*100
free_percent= (free/(free+paid))*100
print(f"There are {round(paid_percent)}% paid app. ")
print(f"There are {round(free_percent)}% free app. ")
```
